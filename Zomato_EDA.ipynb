```python
df['rating'] = (
    df['rating']
    .astype(str)
    .str.split('/')
    .str[0]
)

df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
vote = (
    df.groupby('restaurants', as_index=False)
      .agg({
          'votes': 'mean',
          'rating': 'mean'
      })
      .sort_values(by=['rating', 'votes'], ascending=False)
      .head(10)
)
```

Best Location by Votes

(Relation between Location and Votes)

```python
fig = px.bar(
    x=loc_vote.index,
    y=loc_vote.values,
    color=loc_vote.index,
    title='Best location by votes',
    labels={'x': 'Location', 'y': 'Votes'}
)
fig.update_layout(
    width=700,   # 👈 change this
    height=450   # 👈 and this
)


fig.show()
```

Best Location by Rating

(Relation between Location and Rating)

```python
fig=px.bar(y=vote['rating'],
           x=vote.index,
           color=vote.index # Each bar is colored differently
           ,title='Best location by rating',labels={'x':'Location','y':'Rating'} # changes x and y label
          )
fig.update_layout(
    width=700,   # 👈 change this
    height=400   # 👈 and this
)

fig.show()
```

•	Most famous restaurant chains in Bengaluru

```python
fam = df['restaurants'].value_counts()
fam
```

```python
top_chains_df = fam.reset_index()
top_chains_df.columns = ['Restaurant_Name', 'Outlet_Count']
top_chains_df.columns
```

```python
top_chains_df = top_chains_df.sort_values(
    by='Outlet_Count',
    ascending=False
)
```

```python
import plotly.express as px

fig = px.bar(
    top_chains_df,
    x='Restaurant_Name',
    y='Outlet_Count',
    color='Restaurant_Name',
    orientation='h',
    title='Most Famous Restaurant Chains in Bengaluru'
)
fig.show()
```
